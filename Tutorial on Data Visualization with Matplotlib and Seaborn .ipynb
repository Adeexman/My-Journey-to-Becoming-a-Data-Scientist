```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

### Line Plot

```python
Years = [2012, 2013, 2014, 2015, 2016]
No_of_Produce_per_year = [2, 4, 8, 10, 12]
```

```python
#To plot and assign DAta to x&y axis
plt.plot(Years, No_of_Produce_per_year);
```

### To Label Each Axis

```python
plt.plot(Years, No_of_Produce_per_year)
plt.xlabel('year')
plt.ylabel('Produce_per_year');
```

### Plotting Multiple Lines

```python
Years = [2012, 2013, 2014, 2015, 2016]
No_of_Produce_per_year = [2, 4, 8, 10, 12]
No_Planted=[2, 1, 2, 4, 6]
```

```python
plt.plot(Years, No_of_Produce_per_year)
plt.plot(Years, No_Planted)
```

### Adding Chart Title and Legend

```python
plt.plot(Years, No_of_Produce_per_year)
plt.plot(Years, No_Planted)

plt.title("Crop_Yields")
plt.legend(["No_of_Produce_per_year", "No_Planted"])
```

### LIne Markers

```python
#The line markers help to trace the change in line with respect to the axis
plt.plot(Years, No_of_Produce_per_year, marker="o")
plt.plot(Years, No_Planted, marker="x")

plt.title("Crop_Yields")
plt.legend(["No_of_Produce_per_year", "No_Planted"])

```

### Styling Lines and Markers

```python
#color or c : set the color of line
#linstyle or ls: choose the solid or dashed line
#linewidth or lw: sets the width of a line
#markersize or ms: sets the size of a marker
#markeredgecolor or mec: sets the edge color of a marker
#markeredgewidth or mew: sets the edge width of a marker
#markerfacecolor or mfc: sets the face color of a marker
#alpha: opacity of the plot
```

```python
#The line markers help to trace the change in line with respect to the axis
plt.plot(Years, No_of_Produce_per_year, marker="o", c="red", ls="-", lw=2, ms=8, mec="green", mew=3, mfc="white", alpha=1)
plt.plot(Years, No_Planted, marker="x", c="black", ls="--", lw=4, ms=10, mec="blue", mew=6, mfc="white", alpha=1)

plt.title("Crop_Yields")
plt.legend(["No_of_Produce_per_year", "No_Planted"])

```

```python
#The line markers help to trace the change in line with respect to the axis
plt.plot(Years, No_of_Produce_per_year, marker="s", c="red", ls="-", lw=2, ms=8, mec="green", mew=3, mfc="white", alpha=1)
plt.plot(Years, No_Planted, marker="v", c="black", ls="--", lw=4, ms=10, mec="blue", mew=6, mfc="white", alpha=1)

plt.title("Crop_Yields")
plt.legend(["No_of_Produce_per_year", "No_Planted"])

```
